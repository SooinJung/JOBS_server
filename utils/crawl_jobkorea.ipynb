```python
def remove_question_prefix(text):
    if isinstance(text, str) and text.startswith('질문Q'):
        return text[5:]  # 맨 앞 5글자를 제거
    return text


df = df.applymap(remove_question_prefix)
df
```

```python
def remove_answer_prefix(text):
    if isinstance(text, str) and text.startswith('답변"'):
        return text.replace('답변"', '', 1)
    return text

df = df.applymap(remove_answer_prefix)
df
```

```python
import numpy as np
def remove_notice_data(row):
    notice_found = False
    for idx, val in enumerate(row):
        if isinstance(val, str) and '공지사항' in val:
            notice_found = True
        if notice_found:
            row[idx] = np.nan
    return row

df = df.apply(remove_notice_data, axis=1)
df
```

```python
import pandas as pd
import re

# 문자열에서 "답변"을 맨 앞에서 제거하고, "보기"를 맨 뒤에서 제거하며, "글자수 자 byte" 형식을 제거하는 함수
def clean_text(text):
    if isinstance(text, str):
        # "답변"으로 시작하는 경우 제거
        if text.startswith('답변'):
            text = text[2:]
        # "보기"로 끝나는 경우 제거
        if text.endswith('보기'):
            text = text[:-2]
        # "글자수 nnn자 nnnByte" 형식의 문자열 제거
        text = re.sub(r'글자수\d+자\d+Byte', '', text)
    return text

# 데이터프레임의 모든 셀에 함수를 적용
df = df.applymap(clean_text)
df
```
